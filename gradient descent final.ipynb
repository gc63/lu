```python
import matplotlib.pyplot as plt
```

```python
def function(x):
    return (x+3)**2

def gradient(x):
    return 2*(x+3)
```

```python
def gradient_descent(start,rate,iteration):
    x=start
    x_values=[x]
    for i in range(iteration):
        grad=gradient(x)
        x=x-rate*grad
        x_values.append(x)
        if abs(grad)<1e-6:
            break
        return x,x_values
```

```python
start=2
rate=0.1
iteration=100
final_x,x_values=gradient_descent(start,rate,iteration)
print(f"local minima x: {final_x:.6f}")
```

```python
x_range=range(-10,5)
y_value=[function(x) for x in x_range]
```

```python
plt.plot(x_range,y_value,label='y=(x+3)^2')
plt.scatter(x_values,[function(x) for x in x_values],label='Gradient Descent')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
